```python
import numpy as np
import matplotlib.pyplot as 


###animate####!!
#------------------------#
def f(A,w,t):
	return A*np.cos(w*t)
#------------------------#
def theta_to_xy(theta,l):
    x=l*np.sin(theta)
    y=-l*np.cos(theta)
    return x,y
#------------------------#
A=1
g=9.8 
l=1.0 #m
w=np.sqrt(g/l)
t=np.arange(0,10*np.pi,0.1)
points= theta_to_xy(f(A,w,t),l)
plt.plot(points[0], points[1], ".")
plt.show()

```
